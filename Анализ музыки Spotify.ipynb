```python
import pandas as pd
df= pd.read_csv('data_sp.csv')
pd.options.display.max_columns = 26
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
import matplotlib.pyplot as plt
```

```python
columns=['artists', 'danceability', 'duration_ms', 'explicit','name','duration_ms', 'speechiness', 'year']
```

```python
df = df[columns]
```

```python
df=df.sort_values(by=['popularity'],ascending=False)
df.head(50)
```

```python
df_year = df[(df["year"]>=1980) & (df["year"]<=2021)].copy()
df_year["year"] = df_year["year"].astype(int)
plt.figure(figsize=(20,7))
sns.countplot(data= df_year,x="year")
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation= -35)
sns.barplot(data= df_year,x="popularity", y= "acousticness",estimator=median, palette='hls')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation= -35)
sns.barplot(data= df_year,x="popularity", y= "instrumentalness",estimator=median, palette='hls')
plt.show()
```

```python
df_year['artists'].value_counts().head(17).plot.bar(figsize=(27,8),rot=0, title="Top 17 models")
plt.show()
```
